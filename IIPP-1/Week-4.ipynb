```python
# Lecture-3 (Interactive Drawing);

# This lecture essentailly combines the above two lecture to pring the user inout in terms of dollar and cents!
# interactive application to convert a float in dollars and cents

import simplegui

# define global value
value = 3.12

# Handle single quantity
def convert_units(val, name):
    result = str(val) + " " + name
    if val > 1:
        result = result + "s"
    return result
        
# convert xx.yy to xx dollars and yy cents
def convert(val):
    # Split into dollars and cents
    dollars = int(val)
    cents = int(round(100 * (val - dollars)))

    # Convert to strings
    dollars_string = convert_units(dollars, "dollar")
    cents_string = convert_units(cents, "cent")

    # return composite string
    if dollars == 0 and cents == 0:
        return "Broke!"
    elif dollars == 0:
        return cents_string
    elif cents == 0:
        return dollars_string
    else:
        return dollars_string + " and " + cents_string
    
# define draw handler
def draw(canvas):
    canvas.draw_text(convert(value), [60, 110], 24, "White")

# define an input field handler
def input_handler(text):
    global value
    value = float(text)

# create frame
frame = simplegui.create_frame("Converter", 400, 200)

# register event handlers
frame.set_draw_handler(draw)
frame.add_input("Enter value", input_handler, 100)

# start frame
frame.start()
```

```python
# Practice exercise (Challenge, controlling the radius of the ball):

# Move a ball
###################################################
# Student should add code where relevant to the following.
import simplegui 

# Define globals - Constants are capitalized in Python
HEIGHT = 400
WIDTH = 400
RADIUS_INCREMENT = 5
ball_radius = 20

# Draw handler (center, radius, line_width, line_color)
def draw(canvas):
    canvas.draw_circle([WIDTH/2, HEIGHT/2], ball_radius, 12, 'Green')

# Event handlers for buttons
def increase_radius():
    global ball_radius
    ball_radius += RADIUS_INCREMENT
    
def decrease_radius():
    global ball_radius
    if ball_radius>RADIUS_INCREMENT:
        ball_radius -= RADIUS_INCREMENT    
    

# Create frame and assign callbacks to event handlers
frame = simplegui.create_frame("Ball control", WIDTH, HEIGHT)
frame.set_draw_handler(draw)
frame.add_button("Increase radius", increase_radius)
frame.add_button("Decrease radius", decrease_radius)

# Start the frame animation
frame.start()
```

```python
# Lecture-4 (Timers);

# A "screensaver" program
# First time introducing lists where the text starts from the point

# Import modules
import simplegui, random

# Global state
message = "Python is Fun!"
position = [50, 50]
WIDTH = 500
HEIGHT = 500
interval = 500

# Handler for text box
def update(text):
    global message
    message = text
    
# Handler for timer
def tick():
    x = random.randrange(0, WIDTH)
    y = random.randrange(0, HEIGHT)
    position[0] = x
    position[1] = y

# Handler to draw on canvas
def draw(canvas):
    canvas.draw_text(message, position, 36, "Red")

# Create a frame 
frame = simplegui.create_frame("Home", width, height)

# Register event handlers
frame.set_draw_handler(draw)
text = frame.add_input("Message:", update, 150)
timer = simplegui.create_timer(interval, tick)

# Start the frame animation
frame.start()
timer.start()
```

```python
# Practice Exercises 

# Starting and stopping the timer is important!
# Counter with buttons
###################################################
# Student should add code where relevant to the following:

import simplegui 

counter = 0

# Timer handler
def tick():
    global counter
    print counter
    counter += 1
    
# Event handlers for buttons    
def start():
    timer.start()
    
def stop():
    timer.stop()

def reset():
    global counter
    counter = 0
        
# Create frame and timer
frame = simplegui.create_frame("Counter with buttons", 200, 200)
frame.add_button("Start", start, 100)
frame.add_button("Stop", stop, 100)
frame.add_button("Reset", reset, 100)
timer = simplegui.create_timer(1000, tick)

# Start timer
timer.start()


# Challenge exercise:

# Reflex tester/ Counts the number of seconds between two consecutive clicks!
###################################################
# Student should add code where relevant to the following.

import simplegui 

total_ticks = 0
first_click = True

# Timer handler
def tick():
    global total_ticks
    total_ticks += 1
    
# Button handler
def click():
    global first_click, total_ticks
    if first_click == True:
        timer.start()
        first_click = False
    else:
        timer.stop()
        print "the time elapsed in seconds is ", total_ticks*0.01
        first_click = True
        total_ticks = 0
        
# Create frame and timer
frame = simplegui.create_frame("Counter with buttons", 200, 200)
frame.add_button("Click me", click, 200)
timer = simplegui.create_timer(10, tick)

# Start timer
frame.start()
```
