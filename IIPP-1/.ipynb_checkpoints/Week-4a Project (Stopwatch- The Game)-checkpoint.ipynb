```python
# template for "Stopwatch: The Game"
import simplegui

# define global variables
counts = 0 
width = 500
height = 500
stop_times = 0
success = 0 
state = False

# define helper function format that converts time
# in tenths of seconds into formatted string A:BC.D
def format(t):
    D = str(t%10)
    t = t//10
    temp = t%60
    if temp < 10:
        BC = "0" + str(temp) 
    else:
        BC = str(temp)
    t = t//60
    A = str(t)
    return A + ":" + BC + "." + D

message = format(counts)

# define event handlers for buttons; "Start", "Stop", "Reset"
def start():
    global state
    timer.start()
    state = True
    
def stop():
    global stop_times, success, state
    timer.stop()
    
    if state == True:   
        stop_times += 1
        state = False
        
        if counts%10 == 0:
            success += 1 
        
def reset():   
    global counts, stop_times, success, message, state
    counts = 0
    stop_times = 0
    success = 0
    message = format(counts)
    state = False
    
# define event handler for timer with 0.1 sec interval
def tick():
    global counts, message
    counts += 1
    message = format(counts)
    
# define draw handler
def draw(canvas):
    canvas.draw_text(message, [width/2, height/2], 36, "Red")
    canvas.draw_text(str(success) + "/" + str(stop_times), [0.8*width, 0.1*height], 36, "Red")

# create frame
frame = simplegui.create_frame("Timer", width, height)

# register event handlers
frame.set_draw_handler(draw)
frame.add_button("Start", start, 100)
frame.add_button("Stop", stop, 100)
frame.add_button("Reset", reset, 100)
timer = simplegui.create_timer(100, tick)

# start frame
frame.start()

# Please remember to review the grading rubric
```
