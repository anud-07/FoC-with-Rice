```python
# Lecture-3 (Motion)

# Motion that directly changes the position of the circle to show motion;
# Ball motion with an implicit timer
import simplegui

# Initialize globals
WIDTH = 600
HEIGHT = 400
BALL_RADIUS = 20

ball_pos = [WIDTH / 2, HEIGHT / 2]
vel = [0, 1] # pixels per update (1/60 seconds)

# define event handlers
def draw(canvas):
    # Update ball position
    ball_pos[0] += vel[0]
    ball_pos[1] += vel[1]

    # Draw ball
    canvas.draw_circle(ball_pos, BALL_RADIUS, 2, "Red", "White")

# create frame
frame = simplegui.create_frame("Motion", WIDTH, HEIGHT)

# register event handlers
frame.set_draw_handler(draw)

# start frame
frame.start()
```

```python
# Lecture-4 (Collisions and Reflections)

# Some simple conditions for finding positions and velocities after Collisions and Reflections!
import simplegui

# Initialize globals
WIDTH = 600
HEIGHT = 400
BALL_RADIUS = 20

ball_pos = [WIDTH / 2, HEIGHT / 2]
vel = [-60.0 / 60.0,  60.0 / 60.0]

# define event handlers
def draw(canvas):
    # Update ball position
    ball_pos[0] += vel[0]
    ball_pos[1] += vel[1]
    
    # collide and reflect off of left hand side of canvas
    if ball_pos[0] <= BALL_RADIUS:
        vel[0] = - vel[0]

    
    # Draw ball
    canvas.draw_circle(ball_pos, BALL_RADIUS, 2, "Red", "White")

# create frame
frame = simplegui.create_frame("Ball physics", WIDTH, HEIGHT)

# register event handlers
frame.set_draw_handler(draw)

# start frame
frame.start()
```

```python
# Some optional practice exercises after lecture!

# List reference problem: IMPORTANT!
###################################################
# Student should enter code below
a = [5, 3, 1, -1, -3, 5]
b = a
b[0] = 0
print a
print b
# Explanation:
# The assignment b = a created a second reference to a.
# Setting b[0] = 0 also mutated the list that both 
# a and b reference.  As a result, a[0] == 0.

# How to avoid the other problem?: IMPORTANT!
###################################################
# Student should enter code below
a = [5, 3, 1, -1, -3, 5]
b = list(a)
b[0] = 0
print a
print b
# Explanation:
# The assignment b = list(a) created a copy of the list a.
# Setting b[0] = 0 only mutated the copy of the list, not
# the original list.  As a result, a[0] == 5.
```

```python
# Lecture-6 (Visualizing Lists and Mutation):

###################################
# Mutation vs. assignment

################
# Look alike, but different
a = [4, 5, 6]
b = [4, 5, 6]
print("Original a and b:", a, b)
print("Are they same thing?", a is b) # Whether they are same can be confirmed from the "IS" condition or operator! 
a[1] = 20
print("New a and b:", a, b)
print(" ")

################
# Aliased
c = [4, 5, 6]
d = c
print("Original c and d:", c, d)
print("Are they same thing?", c is d) # Whether they are same can be confirmed from the "IS" condition or operator! 
c[1] = 20
print("New a and b:", c, d)
print(" ")

################
# Copied

e = [4, 5, 6]
f = list(e)
print("Original e and f:", e, f)
print("Are they same thing?", e is f) # Whether they are same can be confirmed from the "IS" condition or operator! 
e[1] = 20
print("New a and b:", e, f)
print(" ")

###################################
# the effect of MUTATION and ASSIGNMENT interact with global and local variables:
# visualize using the VIZ mode in codeskulptor!

a = [4, 5, 6]

def mutate_part(x):
    # performs a global "a" update mutation;
    # mutation operation!
    a[1] = x

def assign_whole(x):
    # performs a local "a" update, does not afffect the "a" outside;
    # assignment operation!
    a = x

def assign_whole_global(x):
     # performs a global "a" update replacing any other "a" variable;
    global a
    a = x

mutate_part(100)
print(a)
     
assign_whole(200)
print(a)

assign_whole_global(300)
print(a)

# What are the key take-aways?
# 1. for assignment, the "a" variable is assumed to a local varible until explicitly defined as global using keyword!
# 2. for mutation, the "a" variable is assumed to be a global variable 

# The above observation begs the question "what is mutable" and "what is im-mutatble"?
# 1. to answer this question, we know that the result after applying the "IS" operator answers this question!
```

```python
# Lecture-7 (Programming Tips - 4):

# The fact that lists are mutable, we need similar datatypes that are NOT mutable!
###################################
# Lists (mutable) vs. tuples (immutable)

print([4, 5, 6])
print((4, 5, 6))
print(type([4, 5, 6]))
print(type((4, 5, 6)))

a = [4, 5, 6]
a[1] = 100
print(a)

b = (4, 5, 6)
b[1] = 100
print(b)

# However, both lists and tuple support the same non-mutatble operations such as you can LOOP on tuples!
# Generally used when the data should not be accidently changed!
```
