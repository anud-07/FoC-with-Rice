```python
# Template for "Guess the number" mini-project
# Input will come from buttons and an input field
# All output for the game will be printed in the console
import simplegui
import random 

# helper function to start and restart the game
def new_game():
    # initialize global variables used in your code here
    global secret_number
    secret_number = random.randrange(0, des_range)

# define event handlers for control panel
def range100():
    # button that changes the range to [0,100) and starts a new game 
    print "Restarting the game with range upto 100"
    global des_range, max_guesses
    des_range = 100
    max_guesses = 7
    new_game()
    print "You have a total of ", max_guesses, " tries to hit the jackpot"

def range1000():
    # button that changes the range to [0,1000) and starts a new game    
    print "Restarting the game with range upto 1000"    
    global des_range, max_guesses
    des_range = 1000
    max_guesses = 10    
    new_game()
    print "You have a total of ", max_guesses, " tries to hit the jackpot"

def reset():
    if des_range == 100:
        range100()
    else:
        range1000()    
    
def input_guess(guess):
    # main game logic goes here	
    print " " 
    global max_guesses
    max_guesses -= 1

    guess = int(guess)
    print "Guess was ", guess
    
    if guess == secret_number:
        print "You've hit the jackpot!"
        reset()
    else:
        if max_guesses==0:
            print "You lose my friend"
            reset()
        else:    
            if guess > secret_number:
                print "Go lower"
                print "You have a total of ", max_guesses, " more tries to hit the jackpot"
            else:
                print "Go higher"
                print "You have a total of ", max_guesses, " more tries to hit the jackpot"            
        
# create frame
frame = simplegui.create_frame("Guess the number", 300, 300)
frame.add_input("Enter Your Guess", input_guess, 100)
frame.add_button("Range is [0,100)", range100, 100)
frame.add_button("Range is [0,1000)", range1000, 100)

frame.start()

# call new_game 
# new_game()

# always remember to check your completed program against the grading rubric
```
