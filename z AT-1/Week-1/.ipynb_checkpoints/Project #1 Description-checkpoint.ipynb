```python
EX_GRAPH0 = { 0:set([1,2]), 
              1:set(), 
              2:set()}

EX_GRAPH1 = { 0:set([1,4,5]), 
              1:set([2,6]), 
              2:set([3]), 
              3:set([0]), 
              4:set([1]), 
              5:set([2]), 
              6:set()}

EX_GRAPH2 = { 0:set([1,4,5]), 
              1:set([2,6]), 
              2:set([3,7]), 
              3:set([7]), 
              4:set([1]), 
              5:set([2]), 
              6:set(), 
              7:set([3]), 
              8:set([1,2]),
              9:set([0,3,4,5,6,7])}

def make_complete_graph(num_nodes):
    """
    input: number of nodes
    returns a dictionary representation of a complete graph
    """
    graph_dict = {}
    
    if num_nodes<=0:
        return graph_dict 
    
    neig = set(range(num_nodes))
    for node in range(num_nodes):
        graph_dict[node] = neig.difference(set([node]))
    
    return graph_dict
    
# an example    
# num_nodes = 5
# print make_complete_graph(num_nodes)    

#print ""


# Computing in degree
def compute_in_degrees(digraph):
    """
    input: dictionary representation of a directed graph
    returns the in degree of each node
    """
    in_degree_dict = {}
    if len(digraph)==0:
        return in_degree_dict
    
    for keys in digraph:
        in_degree_dict[keys] = 0    
 
    for dummy_keys, dummy_values in digraph.items():
        for neighbor in dummy_values:
            in_degree_dict[neighbor] += 1	    
                  
    return in_degree_dict

GRAPH4 = {"dog": set(["cat"]),
          "cat": set(["dog"]),
          "monkey": set(["banana"]),
          "banana": set([])}
          
# an example
print compute_in_degrees(GRAPH4)    


# print ""


# Computing in degree distribution
def in_degree_distribution(digraph):
    """
    input: dictionary representation of a directed graph
    returns in-degree distribution where keys correspond to in-degrees of nodes in the graph
    and values are the number of nodes with that in-degree
    """    
    in_degree_dict = compute_in_degrees(digraph)
    
    in_degree_dist_dict = {}
    
    for dummy_keys, dummy_values in in_degree_dict.items():
            if dummy_values not in in_degree_dist_dict:
                in_degree_dist_dict[dummy_values] = 0
            in_degree_dist_dict[dummy_values] += 1    
            
    return in_degree_dist_dict      
                
print in_degree_distribution(GRAPH4)                
```
