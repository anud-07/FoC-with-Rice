```python
https://py2.codeskulptor.org/#user48_4oEWQFc8Zo_8.py

# Implementation of card game - Memory
import simplegui, random
deck, exposed, history, turns = [], [], [], 0

# helper function to initialize globals
def new_game():
    global deck, exposed, state, turns, history
    deck1, deck2 = range(1, 9), range(1, 9)
    deck = deck1+deck2
    #random.shuffle(deck)
    exposed = [False for iter in range(len(deck))] 
    state, turns = 0, 0
    label.set_text("Turns = " + str(turns))
    history=[]
    
# define event handlers
def mouseclick(pos):
    global exposed, state, turns, history
      
    click = list(pos) # x-coordinate           
        
    # add game state logic here
    if state==0:
        state=1   # 1-card exposed
        exposed[click[0]//50] = True
        history.append(click[0]//50)
        
    elif state==1: # 2-cards exposed
        if exposed[click[0]//50] == False:
            exposed[click[0]//50] = True
            history.append(click[0]//50)      
            state=2
            turns += 1
            label.set_text("Turns = " + str(turns))

    else: 
        if exposed[click[0]//50] == False:
            if not deck[history[-1]] == deck[history[-2]]:
                exposed[history[-1]] = False
                exposed[history[-2]] = False 
            history = []   
            state=1  
            exposed[click[0]//50] = True
            history.append(click[0]//50) 
                            
# cards are logically 50x100 pixels in size    
def draw(canvas):
    for idx in range(len(deck)): 
        if exposed[idx]: 
            canvas.draw_text(str(deck[idx]), (25*(2*idx+1), 60), 30, 'White')
        else:
            canvas.draw_polygon([[25*(2*idx+1)-25, 0], [25*(2*idx+1)+25, 0], [25*(2*idx+1)+25, 100], [25*(2*idx+1)-25, 100]], 5,  'Black', 'Green')
        
# create frame and add a button and labels
frame = simplegui.create_frame("Memory", 800, 100)
frame.add_button("Reset", new_game)
label = frame.add_label("Turns = 0")

# register event handlers
frame.set_mouseclick_handler(mouseclick)
frame.set_draw_handler(draw)

# get things rolling
new_game()
frame.start()

# Always remember to review the grading rubric
```
